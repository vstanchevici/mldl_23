```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

%matplotlib inline

data = pd.read_csv('../../data/data_flats.csv', sep=";")

#data['sub_area'].info()
#data['price_doc'].info()
lef_price = data.price_doc[data.sub_area == 'Lefortovo']
per_price = data.price_doc[data.sub_area == 'Perovo']
bas_price = data.price_doc[data.sub_area == 'Basmannoe']
bog_price = data.price_doc[data.sub_area == 'Bogorodskoe']

plt.boxplot(lef_price)
print(lef_price.median())
print(lef_price.max()-lef_price.min())
plt.show()

plt.boxplot(per_price)
print(per_price.median())
print(per_price.max()-per_price.min())
plt.show()

plt.boxplot(bas_price)
print(bas_price.median())
print(bas_price.max()-bas_price.min())
plt.show()

plt.boxplot(bog_price)
print(bog_price.median())
print(bog_price.max()-bog_price.min())
plt.show()
```
